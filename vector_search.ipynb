```python
# 1. Khai báo thư viện
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

# 2. Đọc file vật lý "tagged_description.txt" từ ổ cứng
raw_documents = TextLoader("tagged_description.txt").load()

# 3. Băm file thành nhiều đoạn nhỏ và LƯU VÀO BIẾN `documents`
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)

# 4. Tải mô hình nhúng miễn phí
hf_embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

```

```python
documents[0]
```

```python
import os

# 5. Đưa biến `documents` vào ChromaDB (Lúc này máy đã hiểu documents là gì)
db_books = Chroma.from_documents(
    documents,
    embedding=hf_embeddings
)
print("Thành công!")
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
